```python
import pandas as pd
import bqplot
import json
import pandas as pd 
import bqplot.pyplot as plt
import re
import numpy as np
import random
from functools import reduce
import seaborn as sns
from ast import literal_eval
# import matplotlib.pyplot as plt
from matplotlib.colors import Normalize
```

```python
df = pd.read_csv('v1.25-2023-05-11-ror-data.csv',encoding='iso-8859-1')
df[:100]
```

```python
# Chart 1: Bar chart of the count distribution of column 'types'

fig = plt.figure(title="Count Distribution of Types")
types_counts = df['types'].value_counts()
x = list(types_counts.index)
y = types_counts.values
plt.xlabel('Types')
plt.ylabel('Count')
bar = plt.bar(x, y)
fig
```

```python
# Chart 2: Bar chart of the count distribution of column 'status'

fig = plt.figure(title="Count Distribution of Status")
types_counts = df['status'].value_counts()
x = list(types_counts.index)
y = types_counts.values
plt.xlabel('Status')
plt.ylabel('Count')
bar = plt.bar(x, y, colors=["salmon"])
fig
```

```python
# Chart 3: Bar chart of the count distribution of 'acronyms' based on has 'Acronyms' and no acronyms rows

acronyms_counts = df['acronyms'].notna().value_counts()
acronyms_chart = plt.figure(title='Count Distribution of Acronyms in Names')
x = list(acronyms_counts.index.astype(str))
y = acronyms_counts.values
acronyms_bar = plt.bar(x, y, color=["red"])
plt.xlabel('Has Acronyms')
plt.ylabel('Count')
plt.show()
```
